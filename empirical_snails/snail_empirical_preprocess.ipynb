```python
import numpy as np
import os

outdir = "snail_fasta/"
if not os.path.exists(outdir):
    os.mkdir(outdir)

fasta_file = "Naesiotus_mtDNA_clean.fas"
with open(fasta_file) as infile:
    dat = zip(infile, infile)
print(dat[0])
```

```python
## All species are single island endemics, so island code and species name are enough to differentiate them
for samp, seq in dat:
    name = samp.split("_")[0].split(">")[1] + "_" + samp.split("_")[2]
    print(name)
    with open(outdir+name+".fasta", 'a') as outfile:
        outfile.write(samp)
        outfile.write(seq)
```
