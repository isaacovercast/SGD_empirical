Investigating suburban micromoth diversity using DNA barcoding of malaise trap samples

```python
import numpy as np
import os

outdir = "micromoth_fasta/"
if not os.path.exists(outdir):
    os.mkdir(outdir)

fasta_file = "micromoth.fasta"
with open(fasta_file) as infile:
    dat = zip(infile, infile)
print(dat[0])
```

```python
for samp, seq in dat:
    name = samp.split("|")[1].replace(" ", "_")
    with open(outdir+name+".fasta", 'a') as outfile:
        outfile.write(samp)
        outfile.write(seq)
```
